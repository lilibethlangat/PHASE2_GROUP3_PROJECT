```python
# Load the datasets
budget_df = pd.read_csv("movie_budget_data.csv")
merged_movies_df = pd.read_csv("merged_movies_data.csv")

# 1. Prepare Merge Keys (Standardize Title Columns)
# Create the standardized merge key 'title_clean' for combined_movie_data
# Ensure 'title_clean' is standardized in merged_movies_df

budget_df['title_clean'] = budget_df['title'].astype(str).str.lower().str.strip()
merged_movies_df['title_clean'] = merged_movies_df['title_clean'].astype(str).str.lower().str.strip()

# --- 2. Full Outer Join ---
# Combines all rows from both DataFrames, filling missing values with NaN.
full_outer_join_df = pd.merge(
    budget_df,
    merged_movies_df,
    on='title_clean',
    how='outer',       # Specifies a Full Outer Join
    suffixes=('_box_office', '_tmdb_imdb'),
    indicator=True     # Adds a column to show the source of the row
)

print(full_outer_join_df.shape)
print(full_outer_join_df['_merge'].value_counts())
```
